# Cryptocurrency and stock analysis 

- <h3> Importing the libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf
```

- <h3> Getting the data using yfinance

```python
cryptocurrencies = ['DOGE-USD','BTC-USD','ETH-USD','XRP-USD']
```

```python
data = yf.download(cryptocurrencies,start='2020-01-01',end='2022-03-30')
```

- <h3> Displaying the datafrmae

```python
data.head()
```

```python
adj_close = data['Adj Close']
adj_close.head()
```

- <h3> Plotting the adjusted prices of BTC,ETH,DOGE,XRP

```python
fig,axis=plt.subplots(2,2,figsize=(17,7),gridspec_kw ={'wspace':0.1,'hspace':0.5})

axis[0,0].plot(adj_close['DOGE-USD'])
axis[0,0].set_title('DOGE')

axis[0,1].plot(adj_close['BTC-USD'])
axis[0,1].set_title('BTC')

axis[1,0].plot(adj_close['ETH-USD'])
axis[1,0].set_title('ETH')

axis[1,1].plot(adj_close['XRP-USD'])
axis[1,1].set_title('XRP')
plt.show()
```

- <h3> Calculating the percentage change returns

```python
returns = adj_close.pct_change().dropna(axis=0)
returns
```

- <h5> Plotting the percentage change returns

```python
fig,axis=plt.subplots(2,2,figsize=(17,7),gridspec_kw ={'wspace':0.1,'hspace':0.5})

axis[0,0].plot(returns['DOGE-USD'])
axis[0,0].set_title('Returns on DOGE')

axis[0,1].plot(returns['BTC-USD'])
axis[0,1].set_title('Returns on BTC')

axis[1,0].plot(returns['ETH-USD'])
axis[1,0].set_title('Returns on ETH')

axis[1,1].plot(returns['XRP-USD'])
axis[1,1].set_title('Returns on XRP')
plt.show()
```

- <h3> Histogram volatility plots of returns on BTC, ETH, DOGE, XRP

```python
fig,axis=plt.subplots(2,2,figsize=(17,7),gridspec_kw ={'wspace':0.1,'hspace':0.5})

axis[0,0].hist(returns['DOGE-USD'],bins=100, range=(-0.5,0.5))
axis[0,0].set_title('Histogram of returns on DOGE')

axis[0,1].hist(returns['BTC-USD'],bins=100, range=(-0.5,0.5))
axis[0,1].set_title('Histogram of returns on BTC')

axis[1,0].hist(returns['ETH-USD'],bins=100, range=(-0.5,0.5))
axis[1,0].set_title('Histogram of returns on ETH')

axis[1,1].hist(returns['XRP-USD'],bins=100, range=(-0.5,0.5))
axis[1,1].set_title('Histogram of returns on XRP')
plt.show()
```

- <h3> Correlation matrix between BTC, ETH, DOGE and XRP

```python
corr = returns.corr()
sns.heatmap(corr,annot=True,cmap='Greens')
plt.show()
```
